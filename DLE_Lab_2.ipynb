```python
import numpy as np
```

```python
x = np.array([75, 65, 80])
w = np.array([4, 3, 5])
b = -1000
n = 0.01
y = 1
epochs = 5
```

```python
def step(z):
  return z >= 0
```

```python
for _ in range(epochs):
  z = np.dot(x, w) + b
  y_pred = step(z)
  error = y - y_pred

  if error == 0:
    break

  w = w + (n * error * x)
  b = b + n * error
```
